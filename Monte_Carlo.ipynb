## Simple Monte Carlo simulation for a transfer function

```python
import random
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# number of iterations
iterations = 10000

# transfer function definition
def transfer_function(A,B):
    return 20*A+5*B

# Monte Carlo simulation function 
def Monte_Carlo(transfer_function, iterations):
    final_results = []
    for n in range(iterations):
        a = random.gauss(200, 33)
        b = random.gauss(50, 5)
        final_results.append(transfer_function(a,b))
    return final_results
```

```python
k = Monte_Carlo(transfer_function, iterations=iterations)
l = Monte_Carlo(transfer_function, iterations=100)
```

```python
fig = plt.figure(figsize=(10,5))
sns.distplot(k, label="10000 iter")
sns.distplot(l, label="100 iter")
plt.legend()
plt.title('Monte Carlo Simulation');
```
